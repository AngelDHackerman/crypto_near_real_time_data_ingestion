```python
import pandas as pd 
import pyarrow.parquet as pq
```

```python
table = pq.read_table("../../Descargas/day28_hour3_newRaw.parquet")
print(table.schema)         # muestra las columnas y tipos
print(table.to_pandas().head(10))  # primeras 10 filas
```

```python
df = table.to_pandas()
df = df.sort_values(by='asset_id')
df.head(45)
```

```python
print("Filas totales:", len(df))
print("Assets distintos:", df["asset_id"].nunique())
print("Symbols distintos:", df["symbol"].nunique())

# Ver supplies para sanity check
print(df[["asset_id", "symbol", "circulating_supply", "max_supply"]].head(20))

# Checar nulos
print(df[["circulating_supply", "max_supply"]].isna().sum())
```
